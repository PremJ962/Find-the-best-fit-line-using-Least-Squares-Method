```python
# NAME: PREM J
# REG NO: 212222080042

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
data = pd.read_csv("FuelConsumption.csv")
data.head()

# Q1: Scatter plot between CYLINDERS vs CO2 EMISSION (green color)
plt.scatter(data['CYLINDERS'], data['CO2EMISSIONS'], color='green')
plt.xlabel('CYLINDERS')
plt.ylabel('CO2 EMISSION')
plt.title('CYLINDERS vs CO2 EMISSION')
plt.show()
```

```python
# Q2: Scatter plot comparing CYLINDERS vs CO2 EMISSION and ENGINESIZE vs CO2 EMISSION using different colors
plt.scatter(data['CYLINDERS'], data['CO2EMISSIONS'], color='blue', label='CYLINDERS')
plt.scatter(data['ENGINESIZE'], data['CO2EMISSIONS'], color='red', label='ENGINESIZE')
plt.xlabel('Feature')
plt.ylabel('CO2 EMISSION')
plt.title('Comparison of CYLINDERS and ENGINESIZE vs CO2 EMISSION')
plt.legend()
plt.show()
```

```python
# Q3: Scatter plot comparing CYLINDERS vs CO2 EMISSION, ENGINESIZE vs CO2 EMISSION, and FUELCONSUMPTION_COMB vs CO2 EMISSION using diff
plt.scatter(data['CYLINDERS'], data['CO2EMISSIONS'], color='blue', label='CYLINDERS')
plt.scatter(data['ENGINESIZE'], data['CO2EMISSIONS'], color='red', label='ENGINESIZE')
plt.scatter(data['FUELCONSUMPTION_COMB'], data['CO2EMISSIONS'], color='green', label='FUELCONSUMPTION_COMB')
plt.xlabel('Feature')
plt.ylabel('CO2 EMISSION')
plt.title('Comparison of Features vs CO2 EMISSION')
plt.legend()
plt.show()
```

```python
# Q4: Train model with independent variable as CYLINDERS and dependent variable as CO2EMISSIONS
X_cylinder = data['CYLINDERS'].values.reshape(-1, 1)
y = data['CO2EMISSIONS'].values
X_train, X_test, y_train, y_test = train_test_split(X_cylinder, y, test_size=0.2, random_state=42)
model_cylinder = LinearRegression()
model_cylinder.fit(X_train, y_train)
```

```python
# Q5: Train model with independent variable as FUELCONSUMPTION_COMB and dependent variable as CO2EMISSIONS
X_fuel = data['FUELCONSUMPTION_COMB'].values.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X_fuel, y, test_size=0.2, random_state=42)
model_fuel = LinearRegression()
model_fuel.fit(X_train, y_train)
```

```python
# Q6: Train models on different train-test ratios and note down their accuracies
ratios = [0.6, 0.7, 0.8, 0.9]
accuracies_cylinder = []
accuracies_fuel = []
```

```python
for ratio in ratios:
 X_train, X_test, y_train, y_test = train_test_split(X_cylinder, y, test_size=(1 - ratio), random_state=42)
 model_cylinder = LinearRegression()
 model_cylinder.fit(X_train, y_train)
 accuracy_cylinder = model_cylinder.score(X_test, y_test)
 accuracies_cylinder.append(accuracy_cylinder)
 X_train, X_test, y_train, y_test = train_test_split(X_fuel, y, test_size=(1 - ratio), random_state=42)
 model_fuel = LinearRegression()
 model_fuel.fit(X_train, y_train)
 accuracy_fuel = model_fuel.score(X_test, y_test)
 accuracies_fuel.append(accuracy_fuel)
print("Accuracy for CYLINDERS model with different train-test ratios:", accuracies_cylinder)
print("Accuracy for FUEL model with different train-test ratios:", accuracies_fuel)
```
